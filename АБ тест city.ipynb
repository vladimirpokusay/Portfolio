```python
df['time_came'] = pd.to_datetime(df['time_came'])
# перевести столбец в дату
df['year']=df['time_came'].dt.year

# достаем год из даты
df
```

```python
# чтобы провести Т тест по двум параметрам
df_msk_2020 = df[(df['city'] == 'Moscow') & (df['year'] == 2020)]
```

```python
s, p = ttest_ind(df_msk_2020[df_msk_2020['id_group']==1]['nflag_purchase'], df_msk_2020[df_msk_2020['id_group']==0]['nflag_purchase'])

print(df_msk_2020[df_msk_2020['id_group']==1]['nflag_purchase'].mean()) # срнеднее в тестовой группе
print(df_msk_2020[df_msk_2020['id_group']==0]['nflag_purchase'].mean()) #срнеднее в контрольной группе

print(s) # 0.66 значит разницы нет
print(p)
```

```python
for i in df['city'].unique():
    for j in df['year'].unique():
        print(i,j)
        s, p = ttest_ind(df[(df['city']==i) & (df['year']==j) & (df['id_group']==1)]['nflag_purchase'], df[(df['city']==i) & (df['year']==j) & (df['id_group']==0)]['nflag_purchase'])
        print(p)
        print(s)
        print()
# различия, где p_valuo < 5% есть только по Владивостоку 2020
```

```python
def p_v_city_year(x, y, alpha = 0.05):
    for i in x.unique():
        for j in y.unique():
            #print(i,j)
            s, p = ttest_ind(df[(x==i) & (y==j) & (df['id_group']==1)]['nflag_purchase'], df[(x==i) & (y==j) & (df['id_group']==0)]['nflag_purchase'])
            if p < alpha:
                print('Распределения не равны в', i, 'и p =', p)
    
```

```python
p_v_city_year(df['city'], df['year'])
```

```python
# функция из решения для расчета красивого Т теста

def ab_test_calc(r1, r2):
    s, p = ttest_ind(r1, r2)
    print('Среднее первого ряда = ', r1.mean())
    print('Среднее второго ряда = ', r2.mean())
    print()
    print('Значение стат. критерия = ', s)
    print('Значение p_value = ', p)
    print()
    return s,p
```

```python
for i in df.city.unique():
    
    df_loc = df[df['city'] == i]
    print(i)
    r1 = df_loc[df_loc['id_group'] == 1]['nflag_purchase']
    r2 = df_loc[df_loc['id_group'] == 0]['nflag_purchase']
    
    ab_test_calc(r1, r2)
    print()
```

```python
# для запуска двойного цикла, теста по двум критериям

for i in df.city.unique():
    for j in df.year.unique():
    
        df_loc = df[(df['city'] == i) & (df['year'] ==j)]
        print(i)
        print(j)
        
        r1 = df_loc[df_loc['id_group'] == 1]['nflag_purchase']
        r2 = df_loc[df_loc['id_group'] == 0]['nflag_purchase']
    
        ab_test_calc(r1, r2)
        print()
```
